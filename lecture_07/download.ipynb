```python
import requests
```

```python
data = requests.get("https://i.ytimg.com/vi/j3GoK90M5aM/maxresdefault.jpg")
with open("cat.jpg", "wb") as f:
    f.write(data.content)
```

```python
data = requests.get("https://www.goodreads.com/quotes")

with open("quote.html", "wt", encoding="utf-8") as f:
    f.write(data.text)
```
