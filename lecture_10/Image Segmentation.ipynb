```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iron = cv2.imread("iron.jpg")
```

```python
iron.shape
```

```python
small = cv2.resize(iron, (iron.shape[1]//4, iron.shape[0]//4))
```

```python
small.shape
```

```python
plt.imshow(small)
```

```python
rgb = cv2.cvtColor(small, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
X = rgb.reshape(-1, 3)
```

```python
model = KMeans(2)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_ / 255
```

```python
plt.imshow(pallets.reshape(1, -1, 3))
```

```python
vals = model.labels_.reshape(rgb.shape[:2])
```

```python
gen = np.zeros(rgb.shape)
```

```python
for row in range(vals.shape[0]):
    for col in range(vals.shape[1]):
        gen[row, col] = pallets[vals[row, col]]
```

```python
plt.imshow(gen)
```
