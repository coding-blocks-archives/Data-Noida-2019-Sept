```python
from flask import Flask, render_template, request
import prediction
```

```python
app = Flask("My-app")
```

```python
@app.route("/")
def hello():
    return render_template("form.template")

@app.route("/submit",  methods = ["post"])
def submit():
    text = request.form["text"]
    return prediction.predict(text)

@app.route("/submit",  methods = ["get"])
def back_login():
    return render_template("form.template")
```

```python
app.run(port=6000)
```
